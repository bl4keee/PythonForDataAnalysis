Import pakietu pandas

```python
import pandas as pd
```

Wczytanie pliku state-abbrevs.csv - wczytanie pełnej nazwy stanu oraz skrótu stanu.

```python
abbrevs = pd.read_csv('state-abbrevs.csv', 
                        sep='|', 
                        skiprows=1, 
                        names=['Pełna nazwa', 'Stan'])
abbrevs
```

Wczytanie pliku state-areas.csv - wczytanie pełnej nazwy stanu i powierzchni.

```python
areas = pd.read_csv('state-areas.csv', sep=';', skiprows=1, names=['Pełna nazwa', 'Km2'])
areas
```

```python
Wczytanie pliku state-population.csv - wczytanie stanu, wieku, roku i liczby ludności.
```

```python
population = pd.read_csv('state-population.csv', 
                                    sep=',',
                                    skiprows=1,
                                    dtype=str,
                                    names=['Stan', 'Wiek', 'Rok', 'Liczba ludności'])
population
```

```python
Scalenie tabel z danymi z plików state-abbrevs oraz state-areas.
```

```python
firstMerge = abbrevs.merge(areas, 
                             how='left', 
                             left_on = ['Pełna nazwa'], 
                             right_on=['Pełna nazwa'])
firstMerge
```

```python
Finalne scalenie tabeli z danymi z pliku state-population z danymi z pierwszego scalenia (z powyżej).
```

```python
secondMerge = population.merge(firstMerge, 
                               how='right', 
                               left_on=['Stan'], 
                               right_on=['Stan'])
secondMerge
```

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
graph = nx.Graph()
```

```python
graph.add_edges_from([('A', 'B'), ('A', 'E'), ('A', 'D'),
                      ('B', 'A'), ('B', 'C'), ('B', 'F'),
                      ('C', 'B'), ('C', 'F'), ('C', 'D'), 
                      ('F', 'C'), ('F', 'B'), ('F', 'D'),
                      ('D', 'F'), ('D', 'C'), ('D', 'A'), ('D', 'E'),
                      ('E', 'A'), ('E', 'D')])
nx.draw(graph, with_labels=True, node_color='red')
```

```python
from operator import itemgetter
betc = nx.betweenness_centrality(graph, normalized=False)
print(betc)
sorted(betc.items(), key=itemgetter(1), reverse=True)
```

```python
pos_spring = nx.spring_layout(graph)
```

```python
nodes = graph.nodes()
```

```python
values = [betc.get(node) for node in nodes]
ec = nx.draw_networkx_edges(graph, pos_spring, alpha=0.2)
nc = nx.draw_networkx_nodes(graph, pos_spring, nodelist=nodes, node_color=values, cmap=plt.cm.viridis_r)
ab = nx.draw_networkx_labels(graph, pos_spring, {n:n for n in nodes}, font_size=10)
plt.colorbar(nc)
plt.axis('off')
plt.show()
```

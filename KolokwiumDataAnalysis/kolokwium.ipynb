```
## ZADANIE 6 - REGRESJA LINIOWA
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```
dane2 = dane.dropna().loc[dane.fp56 > 0]
dane.shape #(1510, 43)
dane2.shape #(1510, 43)

## czyli nie było pustych komórek

dane2.plot.scatter(x = "fp55", y = "wiek")
```

```
model1 = smf.ols(formula = "fp56 ~ wiek", data = dane2)
model1
```

```
model1_fit = model1.fit()
model1_fit
```

```
print(model1_fit.summary())
```

```
model2 = smf.ols(formula = "fp56 ~ wiek + fp55", data = dane2).fit()
print(model2.summary())
```
